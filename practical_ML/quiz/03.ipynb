## Right Answer
### Q1. 
All three are essential components of a typical Convolutional Neural Network (CNN):  
**Convolutional Layer** : Applies filters to input data to extract features like edges, textures, etc.  
**Pooling Layer** : Reduces the spatial size of the feature maps (e.g., using max pooling), which helps reduce computation and control overfitting.  
**Activation Function** : Introduces non-linearity (e.g., ReLU) to the network, enabling it to learn complex patterns.  
So, all three should be selected.

### Q3. 
✅ "A convolutional layer outputs one feature map per filter"
Correct. Each filter (kernel) in a convolutional layer produces its own feature map after sliding over the input.

❌ "A convolutional layer can only output one feature map, no matter how many filters defined"
Incorrect. The number of output feature maps equals the number of filters used in the convolutional layer.

✅ "A feature map highlights the areas in the input that activate its corresponding filter the most"
Correct. A feature map shows where and how strongly the input responds to a specific filter's learned pattern.

❌ "A feature map highlights the areas in the input that activate all filters the most"
Incorrect. A feature map corresponds to one specific filter, not all filters.

### Q4. 
❌ "Stride can be 0"
Incorrect. A stride of 0 would mean no movement at all, which is not valid in convolution. Stride must be ≥ 1.

✅ "Stride is the shift from one receptive field to the next"
Correct. Stride determines how far the filter moves across the input (horizontally and/or vertically) each time it slides.

❌ "Stride is the size of receptive fields"
Incorrect. The receptive field size is determined by the filter/kernel size, not the stride.

### Q5. 
❌ "Vertical stride and horizontal stride should always be the same"
Incorrect. They can be different depending on the architecture and design choice.

✅ "Vertical stride and horizontal stride can be different"
Correct. You can set them independently, e.g., (stride height = 2, stride width = 1).

❌ "Vertical stride always equals the height of the receptive field while horizontal stride always equals the width of the receptive field"
Incorrect. Stride and receptive field size are separate parameters. You can have a 3×3 kernel (receptive field) with stride 1 or 2 or more.

❌ "Vertical stride always equals the width of the receptive field while horizontal stride always equals the height of the receptive field"
Incorrect and reversed. Still confuses stride with receptive field size — they are independent.

### Q6.
✅ Size
Refers to the pooling window (e.g., 2×2). This defines how large the region is for computing the average.

✅ Stride
Defines how much the pooling window moves at each step. For example, stride = 2 means the window shifts by 2 units.

❌ Initial weights
Incorrect. Pooling layers (average or max) do not have weights—unlike convolutional or fully connected layers. They perform fixed mathematical operations (like averaging).

✅ Padding type
Padding (e.g., "valid" or "same") determines how the pooling window deals with borders. It affects the output dimensions.

### Q8. 
✅ "When it is SAME padding, the convolutional layer uses zero padding if necessary. The output size is set to the number of input neurons divided by the stride"
Correct. SAME padding adds zero-padding to maintain the output size close to (or equal to) the input size, especially when stride = 1.

❌ "When it is SAME padding, the convolutional layer does not use zero padding and may ignore some rows and columns at the bottom and right of the input image, depending on the stride"
Incorrect. SAME padding does use zero padding. It doesn’t ignore rows/columns unless stride > 1, and even then, it pads to align.

❌ "When it is VALID padding, the convolutional layer uses zero padding if necessary. The output size is set to the number of input neurons divided by the stride"
Incorrect. VALID padding does not use any padding at all.

✅ "When it is VALID padding, the convolutional layer does not use zero padding and may ignore some rows and columns at the bottom and right of the input image, depending on the stride"
Correct. With VALID padding and stride > 1, some bottom/right pixels may be ignored if the filter doesn’t fit exactly.

### Q9.
✅ "The number of filters must be specified when defining a convolutional layer. It is a hyperparameter."
Correct. The number of filters determines how many feature maps the layer will produce.
It's something you choose manually when designing the model—so it's a hyperparameter, not a learnable one.

❌ "The number of filters in a convolutional layer does not need to be specified. It is a learnable parameter."
Incorrect. The number of filters is not learned during training. Only the weights inside each filter are learned.

❌ "The number of filters can be 0."
Incorrect. A convolutional layer with 0 filters would produce no output, which is invalid.
